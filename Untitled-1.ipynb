download statcast data

```python
# ============================================
# Import libraries
# ============================================

from pybaseball import statcast
import pandas as pd
from tqdm import tqdm

# ============================================
# Years we want
# ============================================

years = [2021, 2022, 2023, 2024, 2025]

all_data = []

# ============================================
# Download each season
# ============================================

for year in tqdm(years):

    print(f"\nDownloading Statcast data for {year}...\n")

    season = statcast(
        start_dt=f"{year}-03-01",
        end_dt=f"{year}-11-30"
    )

    season["season"] = year

    all_data.append(season)

# ============================================
# Combine seasons
# ============================================

statcast_data = pd.concat(all_data, ignore_index=True)

print("Shape of dataset:", statcast_data.shape)

# ============================================
# Save to disk
# ============================================

statcast_data.to_parquet("statcast_2021_2025.parquet")

print("\nSaved dataset to: statcast_2021_2025.parquet")
```
